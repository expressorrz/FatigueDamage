```python
import numpy as np
import torch
import sys
sys.path.append('../')
from utils.util import *
from model import *
import matplotlib.pyplot as plt
from matplotlib import rcParams

configs = {
    "font.size": 15
    }

rcParams.update(configs)

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
train_log = np.load('../log/3/training_log.npy')
val_log = np.load('../log/3/validation_log.npy')

print('train_log:', train_log.shape, 'val_log:', val_log.shape)
```

```python
# plot
plt.figure(figsize=(8, 6))

x_t = np.arange(0, train_log.shape[0])
x_v = np.arange(0, train_log.shape[0] + 10, 10)

plt.plot(x_t, train_log, label='Training Loss', color='b')
plt.plot(x_v, val_log, label='Validation Loss', color='r')

plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend()
plt.tight_layout()
plt.show()
```
