```python
import numpy as np
import os
```

```python
source = [3, 9, 11, 12, 13, 14, 19, 20, 25]

x_list = []
y_list = []

for i in source:
    x = np.load(f'../dataset/A{i:03}/x.npy')
    y = np.load(f'../dataset/A{i:03}/y.npy')
    print(f'A{i:03} x shape: {x.shape}, y shape: {y.shape}')
    x_list.append(x)
    y_list.append(y)

x = np.concatenate(x_list, axis=0)
y = np.concatenate(y_list, axis=0)
```

```python
print(x.shape, y.shape)

data_mean = np.mean(x, axis=(0, 1))
data_std = np.std(x, axis=(0, 1))
data_min = np.min(x, axis=(0, 1, 2, 3))
data_max = np.max(x, axis=(0, 1, 2, 3))

# print("Mean per channel:", data_mean.shape)
# print("Standard deviation per channel:", data_std.shape)
print("Min per channel:", data_min)
print("Max per channel:", data_max)


# np.save('./data_mean.npy', data_mean)
# np.save('./data_std.npy', data_std)
np.save('./data_min.npy', data_min)
np.save('./data_max.npy', data_max)

print(x.shape, y.shape)
```
